```python
merged_df=pd.merge(df,countrydf,on='Country Code',how='left')
```

```python
merged_df['Aggregate rating'].value_counts()
```

```python
merged_df['Rating text'].value_counts()
```

```python
ratings=merged_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

```python
sns.countplot(x='Rating color', data=ratings, palette=['blue','red','orange','yellow','green','green'])
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
```
