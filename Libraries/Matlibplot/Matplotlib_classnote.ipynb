```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
plt.bar(company,revenue)
```

```python
plt.xlabel('Major companies')
plt.ylabel('Revenue')
plt.title('Revenue model for Top Companies')
plt.bar(company,revenue)
plt.bar(company,profit)
plt.show()
```

```python
plt.xlabel('Major companies')
plt.ylabel('Revenue')
plt.title('Revenue model for Top Companies')
plt.bar(company,revenue,label='Revenue')
plt.bar(company,profit,label='Profit')
plt.legend()
plt.show()
```

```python
#To get seperate bars for profit and trevenue in the same chart
#we need to give x position a number then change it back to company name

xpos=np.arange(len(company))  
xpos
```

```python
plt.xlabel('Major companies')
plt.ylabel('Revenue')
plt.title('Revenue model for Top Companies')
plt.bar(xpos,revenue,label='Revenue')
plt.bar(xpos,profit,label='Profit')
plt.legend()
plt.show()
```

```python
plt.xticks(xpos,company)   # changing position values to original variable
plt.xlabel('Major companies')
plt.ylabel('Revenue')
plt.title('Revenue model for Top Companies')
plt.bar(xpos,revenue,label='Revenue')
plt.bar(xpos,profit,label='Profit')
plt.legend()
plt.show()
```

```python
plt.xticks(xpos,company)   # changing position values to original variable
plt.xlabel('Major companies')
plt.ylabel('Revenue')
plt.title('Revenue model for Top Companies')
plt.bar(xpos-0.2,revenue,label='Revenue')
plt.bar(xpos+0.2,profit,label='Profit')  #shifting of bars
plt.legend()
plt.show()
```

```python
plt.xticks(xpos,company)   # changing position values to original variable
plt.xlabel('Major companies')
plt.ylabel('Revenue')
plt.title('Revenue model for Top Companies')
plt.bar(xpos-0.2,revenue,width=0.4,label='Revenue') #shifting of bars
plt.bar(xpos+0.2,profit,width=0.4,label='Profit')  #changing width of the bar default value=0.8
plt.legend()
plt.show()
```

```python
#Histogram

blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]  #Histogram automatically sorts to ascending order
plt.hist(blood_sugar) # by default number of bins is set to 10
```

```python
# the arrays shown in the above are

# 1st one is showing the frequency of the data or value
# 2nd one shows the interval or bin created automattically(we can set the bin also)
```

```python
plt.hist(blood_sugar, rwidth=0.8) # rwidth for the required width of the bar
```

```python
plt.hist(blood_sugar, bins=3, rwidth=0.8, color='red') #to change the number of bins
```

```python
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.8)
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.hist([blood_sugar_men,blood_sugar_women])
```

```python
plt.hist([blood_sugar_men,blood_sugar_women],label=['men','women'])
plt.legend()
```

```python
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.6,label=['men','women'])
plt.legend()
```

```python
plt.xlabel('Sugar level')
plt.ylabel('No.of Patients')
plt.title('Sugar_level_chart')
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.6,label=['men','women'],color=['red','yellow'])
plt.legend()
plt.show()
```

```python
exp_vals=[1400,600,300,410,250]
exp_labels=['Home Rent','Food','phone/Internet','Car','Other Utilities']

plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=1.6)   # if needed add plt.legend() after this line
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=1.2,autopct='%0.2f%%')  # to get decimals in % in the chart
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=1.2,autopct='%0.2f%%',shadow=True,explode=[0,0,0,0,0])  #explode between labels
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=1.2,autopct='%0.2f%%',shadow=True,explode=[0.1,0,0.2,0,0]) #order of explode goes anticlockwise
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=1.2,autopct='%0.2f%%',shadow=True,explode=[0.1,0,0.2,0,0],startangle=45) #startangle rotates whole diagram
#plt.show()    # While saving comment or remove the show(), else it will be lank inside the saved file
plt.savefig('Pie chart new.jpg')  #inside matplotlib directory in the local folder . saving can be in any format, applicable for all types of chart
```
