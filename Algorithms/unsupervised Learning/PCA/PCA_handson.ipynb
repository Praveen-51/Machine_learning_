```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv(r"C:\Users\prave\OneDrive\Desktop\ML Learning\Algorithms\unsupervised Learning\PCA\HR.csv")
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data[data.duplicated()]
```

```python
data.drop_duplicates(inplace=True,ignore_index=True)
```

```python
data.head()
```

```python
data.sales.nunique()
```

```python
data=pd.get_dummies(data)
```

```python
data
```

```python
data.corr()
```

```python
data.head()
```

```python
X=data.drop('left',axis=1).values    # .values makes the X into a numpy array
y=data['left']
```

```python
X.shape
```

```python
scaler=StandardScaler()
scaler.fit_transform(X)
```

```python
pca=PCA().fit(X)
```

```python
(pca.explained_variance_ratio_)*100
```

```python
np.cumsum(pca.explained_variance_ratio_)*100
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_)*100)
plt.xlim(0,20,1)
plt.xlabel('Number of components')
plt.ylabel("Cumulative explaned Features")
```

```python
#test accuracy --- shows variance

#train Accuracy --- shows bias
```

```python
# PCA = 2
from sklearn.svm import SVC
sklearn_pca = PCA(n_components=2)
X_pca = sklearn_pca.fit_transform(X)

print("The Original Shape : ",X.shape)
print("The new PCA shape : ",X_pca.shape)
X_train,X_test,y_train,y_test = train_test_split(X_pca,y,test_size=0.3,random_state=10)
svc_classifier = SVC()
svc_classifier.fit(X_train,y_train)
print("Train Acc :", svc_classifier.score(X_train,y_train))
print("Test Acc :", svc_classifier.score(X_test,y_test))
```
