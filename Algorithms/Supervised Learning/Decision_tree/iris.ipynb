```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

from sklearn.datasets import load_iris

import warnings
warnings.filterwarnings('ignore')
```

```python
iris=load_iris()
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
iris.keys()
```

```python
iris_df = pd.DataFrame(iris.data,columns=iris.feature_names)
iris_df["Species"]=iris.target
iris_df.head()
```

```python
X= iris_df.drop(['Species'],axis=1)
y= iris_df['Species']

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=17)
```

```python
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train,y_train)
```

```python
score=dt_classifier.score(X_train,y_train)
print('R-Squared Train= ',score)

score1=dt_classifier.score(X_test,y_test)
print('R-Squared Test= ',score1)


y_pred=dt_classifier.predict(X_test)

mse=mean_squared_error(y_test,y_pred)
print('mse= ',mse)
print('Rmse= ',mse**(1/2))
```

```python
dt_classifier = DecisionTreeClassifier(criterion='entropy',max_depth=3)
dt_classifier.fit(X_train,y_train)

score=dt_classifier.score(X_train,y_train)
print('R-Squared Train= ',score)

score1=dt_classifier.score(X_test,y_test)
print('R-Squared Test= ',score1)


y_pred=dt_classifier.predict(X_test)

mse=mean_squared_error(y_test,y_pred)
print('mse= ',mse)
print('Rmse= ',mse**(1/2))
```
