```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

import warnings
```

```python
data=pd.read_csv(r"C:\Users\prave\OneDrive\Desktop\ML Learning\Algorithms\Supervised Learning\Decision_tree\diabetes.csv")
data.head()
```

```python
X=data.drop(['Outcome'],axis=1)
y=data['Outcome']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train,y_train)

score=dt_classifier.score(X_train,y_train)
print('R-Squared Train= ',score)

score1=dt_classifier.score(X_test,y_test)
print('R-Squared Test= ',score1)


y_pred=dt_classifier.predict(X_test)

mse=mean_squared_error(y_test,y_pred)
print('mse= ',mse)
print('Rmse= ',mse**(1/2))
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(dt_classifier,filled=True,class_names=True)
plt.show()
```

```python
dt_classifier = DecisionTreeClassifier(criterion='entropy',max_depth=4)
dt_classifier.fit(X_train,y_train)

score=dt_classifier.score(X_train,y_train)
print('R-Squared Train= ',score)

score1=dt_classifier.score(X_test,y_test)
print('R-Squared Test= ',score1)


y_pred=dt_classifier.predict(X_test)

mse=mean_squared_error(y_test,y_pred)
print('mse= ',mse)
print('Rmse= ',mse**(1/2))
```

```python
from sklearn import tree
plt.figure(figsize=(20,10))
tree.plot_tree(dt_classifier,filled=True,class_names=True)
plt.show()
```
